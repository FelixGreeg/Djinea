```python
#conexion tipo 3
# Connection parameters, yours will be different
param_dic = {
    "host"      : "localhost",
    "database"  : "postgres",
    "user"      : "postgres",
    "password"  : "postgres",
    "port"      :  5432
}
def connect(params_dic):
    """ Connect to the PostgreSQL database server """
    conn = None
    try:
        # connect to the PostgreSQL server
        print('Connecting to the PostgreSQL database...')
        conn = pg.connect(**params_dic)
    except (Exception, pg.DatabaseError) as error:
        print(error)
        sys.exit(1) 
    print("Connection successful")
    return conn

def postgresql_to_dataframe(conn, select_query, column_names):
    """
    Tranform a SELECT query into a pandas dataframe
    """
    cursor = conn.cursor()
    try:
        cursor.execute(select_query)
    except (Exception, pg.DatabaseError) as error:
        print("Error: %s" % error)
        cursor.close()
        return 1
    
    # Naturally we get a list of tupples
    tupples = cursor.fetchall()
    cursor.close()
    
    # We just need to turn it into a pandas dataframe
    df = pd.DataFrame(tupples, columns=column_names)
    return df
```

```python
# Connect to the database
conn = connect(param_dic)
column_names = ["CVEMUN", "Municipio","%analfa","%sinprim","%sinsec","%rezago"]
# Execute the "SELECT *" query
data = postgresql_to_dataframe(conn, "SELECT cve_geomun, municipio,panalfa,psinprim,psinsec,prezago FROM rezago", column_names)
data.set_index("CVEMUN",inplace=True)
datos=data
datos.dtypes
datos.head()
```

```python
###join con concat
mun1 = gpd.GeoDataFrame.from_file("SHP/Edomex_2010.shp")
#mun.CLAVEMUN = mun.CLAVEMUN.astype(np.int64)
mun1.set_index("CLAVEMUN",inplace=True)
mun1.head()
joincon=pd.concat([mun1,data], axis=1, join='outer', ignore_index=False,#inner,outer
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)
ax=joincon.plot()
ax.set_title("WGS84 (lat/lon)"); 
joincon.crs
```

```python
#join con merge
# Importar mapa
mun2 = gpd.GeoDataFrame.from_file("SHP/Edomex_2010.shp") #Indicamos la dirección del archivo a cargar en formato shp
#mun.CLAVEMUN = mun.CLAVEMUN.astype(np.int64)
mun2.set_index("CLAVEMUN",inplace=True) #se asigna el index para hacer el join
joinmer = mun2.merge(datos,right_on='CVEMUN', left_index=True,how="inner") #left_on right_on, left_index right_index
ax=joinmer.plot()
ax.set_title("WGS84 (lat/lon)"); 
joinmer.crs
```

```python
# Importar mapa
mun = gpd.GeoDataFrame.from_file("SHP/02_Municipios_2010.shp") #Indicamos la dirección del archivo a cargar en formato shp
mun["CVEGEO"]=mun["CLAVEMUN"]
mun.set_index("CVEGEO",inplace=True) #se asigna el index para hacer el join

# join de shape con dataframe,  tipos de union ‘left’, ‘right’, ‘outer’, ‘inner’

join = mun.merge(df,left_on='CVEGEO', right_index=True,how="inner") #left_on right_on, left_index right_index

ax=join.plot()
ax.set_title("WGS84 (lat/lon)"); 

join.crs
join.dtypes
```

```python
# Cargar un mapa base con contornos de países
Edomex = "SHP/Edomex_2010.shp"
map_Edomex = gpd.read_file(Edomex)
fig,ax= plt.subplots(1,1,figsize=(10,10))
map_Edomex.plot(ax=ax)
plt.show()
```

```python
map_Edomex.head()
```

```python
#Asignando centroides
x0=-99.134801
y0=19.685037
#Configurando el canvas en matplotlib y
#creando un objeto "axes"
fig,ax = plt.subplots(1,1,figsize=(10,8))
(map_Edomex["geometry"]
     .skew(xs=30,ys=20,use_radians=False,origin=(x0,y0))
     .rotate(angle=-20,use_radians=False,origin=(x0,y0))
     .plot(ax=ax))
plt.show()
```

```python
mex_tr = (map_Edomex["geometry"]
          .skew(xs=30,ys=20,use_radians=False,origin=(x0,y0))
          .rotate(angle=-20,use_radians=False,origin=(x0,y0)))
x=(mex_tr
   .centroid
   .apply(lambda p:p.x)
   .values)
y=(mex_tr
   .centroid
   .apply(lambda p:p.y)
   .values)
fig,ax = plt.subplots(1,1,figsize=(14,14))
#Graficando el mapa
mex_tr.plot(ax=ax,
            color="#ffffff",
            edgecolor="#bcbcbc",
            zorder=1)

#Mostrando la posición de las coordenadas
ax.scatter(x=x,
           y=y,
           color="#ff944d")
plt.show()
```
